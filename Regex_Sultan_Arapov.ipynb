```python
import re
import numpy as np
# Extract urls like "http://mydomain.com/link-to-listing". Do not use url itself in regular expression. It should select only 4 links.
html = open("sample.html", "r")
html = html.read()
for url in re.findall(r'http:\/\/\w+\.\w+\/\w+-\w+-..\w+', html):
    print(url)
```

```python
# - Extract latitude value from html.
# - Extract longitude value from html.
# 36.5194999,-4.7743365
for number in re.findall(r'[-\d]*\.\d*,.*\d*\.\d*', html):
    print(number)
```

```python
# Extract all unique IDs from json, example of unique ID: "SV350". 
a = open("sample.json", "r")
a = a.read()
print('There are {} unique objects'.format(len(set(re.findall(r'"([a-z][a-z]\d\d\d)"', a, re.IGNORECASE)))))
for i in set(re.findall(r'"([a-z][a-z]\d\d\d)"', a, re.IGNORECASE)):
        print(i)    
```

```python
# Extract all unique IDs from json, that has "land":"ESP" inside, example of UID: SV350.
print("Pattern 'land':'ESP' appears {} times.".format(len(re.findall(r'"land":"esp"', a, re.IGNORECASE))))
print("Consequently, {} unique IDs with 'land':'ESP' were found".format(len(set(re.findall(r'"([a-z][a-z]\d\d\d)".*?("land":"esp")', a, re.IGNORECASE)))))

print("\nUnique IDs with 'land':'ESP'")

for id, land in set(re.findall(r'"([a-z][a-z]\d\d\d)".*?("land":"esp")', a, re.IGNORECASE)):
    print(id)

     
        
```
